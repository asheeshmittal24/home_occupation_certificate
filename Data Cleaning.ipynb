```python
import pandas as pd
import numpy  as np
import seaborn as sns                  # For data visualization
import matplotlib.pyplot as plt 
import seaborn as sn                   # For plotting graphs
%matplotlib inline
import warnings                        # To ignore any warnings
warnings.filterwarnings("ignore")
```

```python
df= pd.read_csv("Home_Occupation_Certificates.csv")
```

```python
df.head()
```

```python
to_drop=['Application Type','Comments','Expiration date','Location','State']
df.drop(columns=to_drop, inplace=True)
```

```python
df.head()
```

```python
df['Application Number'].is_unique
```

```python
df = df.set_index('Application Number')
df.head()
```

```python
df['Parking'].fillna(df['Parking'].mode()[0], inplace=True)
df['Products sold'].fillna(df['Products sold'].mode()[0], inplace=True)
```

```python
df.head()
```

```python
df['Registered Home'].value_counts()
```

```python
df['Registered Home'].replace('N', 0,inplace=True)
df['Registered Home'].replace('Y', 1,inplace=True)
```

```python
corr = df.corr()
mask = np.array(corr)
mask[np.tril_indices_from(mask)] = False
fig,ax= plt.subplots()
fig.set_size_inches(20,10)
sn.heatmap(corr, mask=mask,vmax=.9, square=True,annot=True, cmap="YlGnBu")
```

Through this we can say that Registered Home does not have any correlation with any of the variable.
